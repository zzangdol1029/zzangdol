# Gradient Decent Method

#### Objective function

$ f(x) = 6x^2 + 9x + 3 $

```python
import numpy as np
import matplotlib.pyplot as plt

# IMPLEMENT HERE objective_function(x):
def objective_function(x):
    # f(x) = 6x^2 + 9x + 3
    return 6*x**2 + 9*x + 3

x = np.arange(-5, 5, 0.1)
y = objective_function(x)


plt.plot(x, y)
plt.show()
```

#### Find minimum x using scikit_learn

```python
from scipy import optimize

# IMPLEMENT HERE
# scipy.optimize.minimize   
result = optimize.minimize(objective_function, x0=0, method='BFGS')

plt.plot(x, y)
plt.plot(result.x, objective_function(result.x), 'ro')
plt.show()
```

#### Find minimum x using graident descent method

```python
import numpy as np

def minimize(func, x0, tol=1e-16, max_iter=10000, alpha=0.01):
    """
          
    
    Parameters:
    func:  
    x0: 
    tol:   ( )
    max_iter:   
    alpha:  (step size)
    """
    x = x0
    iteration = 0
    
    #     
    h = 1e-8
    
    print(f"  : x0 = {x0}, alpha = {alpha}")
    print("=" * 50)
    
    for i in range(max_iter):
        #    
        gradient = (func(x + h) - func(x - h)) / (2 * h)
        
        #   
        x_new = x - alpha * gradient
        
        #   
        if abs(x_new - x) < tol:
            print(f" !  : {i+1}")
            break
            
        x = x_new
        iteration = i + 1
        
        #    ( 10  5)
        if i < 10 or i >= max_iter - 5:
            print(f"Iteration {i+1:4d}: x = {x:8.6f}, gradient = {gradient:8.6f}, f(x) = {func(x):8.6f}")
    
    print("=" * 50)
    print(f" : x = {x:.6f}, f(x) = {func(x):.6f}")
    
    return x

result = minimize(objective_function, x0 = 5, alpha = 0.1)

plt.plot(x, y)
plt.plot(result, objective_function(result), 'ro')
plt.show()
```

#### Local minima Problem




```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function_2(x):
    return x**6 - 5*x**4 + 7*x**2 - 3

x = np.arange(-2, 2, 0.1)
y = objective_function_2(x)


plt.plot(x, y)
plt.show()
```

```python
# IMPLEMENT HERE (x0 = -2)
# Local minima    
result = minimize(objective_function_2, x0=-2, alpha=0.01)

plt.plot(x, y)
plt.plot(result, objective_function_2(result), 'ro')
plt.show()
```

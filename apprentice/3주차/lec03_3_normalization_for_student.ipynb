# Normalization vs Standardization

## DataSet

````python
import numpy as np

weight = np.array([5.9, 8, 8.9, 10.1, 10.9, 13.2, 14.8, 16.7, 19.1, 21.5,
                   24.9, 28.5, 32.3, 35.4, 41.5, 45.2, 51, 57.1, 61.7, 63.4,
                   65.2, 67.4, 68.3, 69.5])
height = np.array([59.1, 66.7, 71.4, 75, 80.1, 87.8, 95.2, 102.3, 109, 115.5,
                    122, 127.8, 133.3, 138, 144.9, 150.7, 158.2, 164.7, 169.2, 170.3,
                    172.5, 172.9, 173.4, 173.8])
````

## Show Data Set

````python
#@title { vertical-output: true}
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.scatter(range(len(weight)), weight, color='r', label='weight', s=10)
plt.scatter(range(len(height)), height, color='b', label='height', s=10)
plt.title("Original Data")
plt.legend()
plt.show()
````

## Normalization

````python
def min_max_scaling(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

weight_normalized = min_max_scaling(weight)
height_normalized = min_max_scaling(height)
````

````python
#@title { vertical-output: true}
plt.figure(figsize=(5, 5))
plt.scatter(range(len(weight_normalized)), weight_normalized, color='r', label='weight', s=10)
plt.scatter(range(len(height_normalized)), height_normalized, color='b', label='height', s=10)
plt.title("Normalization")
plt.legend()
plt.show()
````

## Standardization

````python
def standardization(data):
    mean = np.mean(data)
    std = np.std(data)
    return [(x - mean) / std for x in data]

weight_standardized = standardization(weight)
height_standardized = standardization(height)
````

````python
#@title { vertical-output: true}
plt.figure(figsize=(5, 5))
plt.scatter(range(len(weight_standardized)), weight_standardized, color='r', label='weight', s=10)
plt.scatter(range(len(height_standardized)), height_standardized, color='b', label='height', s=10)
plt.title("Standardization")
plt.legend()
plt.show()
````

````python
#     
import numpy as np

#     
weight_mean = np.mean(weight)
weight_variance = np.var(weight)  # 
weight_std = np.std(weight)       # 

#       
height_mean = np.mean(height)
height_variance = np.var(height)  # 
height_std = np.std(height)       # 

print("===    ===")
print(f" :")
print(f"  : {weight_mean:.2f}kg")
print(f"  : {weight_variance:.2f}kg")
print(f"  : {weight_std:.2f}kg")
print(f"   = {weight_std**2:.2f}kg ( )")

print(f"\n :")
print(f"  : {height_mean:.2f}cm")
print(f"  : {height_variance:.2f}cm")
print(f"  : {height_std:.2f}cm")
print(f"   = {height_std**2:.2f}cm ( )")

#   
def manual_variance(data):
    """   """
    mean = np.mean(data)
    squared_diffs = [(x - mean)**2 for x in data]
    return np.mean(squared_diffs)

def manual_std(data):
    """   """
    return np.sqrt(manual_variance(data))

print(f"\n===    ===")
print(f"  (): {manual_variance(weight):.2f}kg")
print(f"  (): {manual_std(weight):.2f}kg")
print(f"  (): {manual_variance(height):.2f}cm")
print(f"  (): {manual_std(height):.2f}cm")
````

````python
#    
import matplotlib.pyplot as plt

#  
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

#   
ax1.scatter(range(len(weight)), weight, color='red', alpha=0.7, s=50)
ax1.axhline(y=weight_mean, color='red', linestyle='--', linewidth=2, label=f': {weight_mean:.1f}kg')
ax1.fill_between(range(len(weight)), 
                 weight_mean - weight_std, 
                 weight_mean + weight_std, 
                 alpha=0.3, color='red', label=f'1 : {weight_std:.1f}kg')
ax1.set_title('  ')
ax1.set_xlabel(' ')
ax1.set_ylabel(' (kg)')
ax1.legend()
ax1.grid(True, alpha=0.3)

#   
ax2.scatter(range(len(height)), height, color='blue', alpha=0.7, s=50)
ax2.axhline(y=height_mean, color='blue', linestyle='--', linewidth=2, label=f': {height_mean:.1f}cm')
ax2.fill_between(range(len(height)), 
                 height_mean - height_std, 
                 height_mean + height_std, 
                 alpha=0.3, color='blue', label=f'1 : {height_std:.1f}cm')
ax2.set_title('  ')
ax2.set_xlabel(' ')
ax2.set_ylabel(' (cm)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

#   
print("===   ===")
print(f" :")
print(f"  : {weight_mean:.1f}kg")
print(f"  : {weight_std:.1f}kg")
print(f"     {weight_mean-weight_std:.1f}kg ~ {weight_mean+weight_std:.1f}kg  ")

print(f"\n :")
print(f"  : {height_mean:.1f}cm")
print(f"  : {height_std:.1f}cm")
print(f"     {height_mean-height_std:.1f}cm ~ {height_mean+height_std:.1f}cm  ")

print(f"\n===  vs   ===")
print(f" : {weight_variance:.1f}kg ( kg )")
print(f" : {weight_std:.1f}kg (  )")
print("     !")
````

##  vs  -  

### (Standardization) = Z-score 
- ****:  0,  1 
- ****: `( - ) / `
- ****:  = 0,  = 1
- ****:    ,  

### (Normalization) = Min-Max Scaling  
- ****:  0,  1 
- ****: `( - ) / ( - )`
- ****:  = 0,  = 1
- ****:   0~1  

###  
|  |  |  |
|------|--------|--------|
| **** |  = 0 |  = 0 |
| **** |  = 1 |  = 1 |
| **** |  | 0 ~ 1 () |
| ** ** |   |   |


````python
#      
print("===  vs   ===")

#   
print(f" :")
print(f"  : ={np.mean(weight):.2f}kg, ={np.std(weight):.2f}kg, ={weight.min():.1f}~{weight.max():.1f}kg")
print(f"  : ={np.mean(height):.2f}cm, ={np.std(height):.2f}cm, ={height.min():.1f}~{height.max():.1f}cm")

print(f"\n  (Min-Max Scaling):")
print(f"  : ={np.mean(weight_normalized):.3f}, ={np.std(weight_normalized):.3f}, ={min(weight_normalized):.3f}~{max(weight_normalized):.3f}")
print(f"  : ={np.mean(height_normalized):.3f}, ={np.std(height_normalized):.3f}, ={min(height_normalized):.3f}~{max(height_normalized):.3f}")

print(f"\n  (Z-score):")
print(f"  : ={np.mean(weight_standardized):.3f}, ={np.std(weight_standardized):.3f}, ={min(weight_standardized):.3f}~{max(weight_standardized):.3f}")
print(f"  : ={np.mean(height_standardized):.3f}, ={np.std(height_standardized):.3f}, ={min(height_standardized):.3f}~{max(height_standardized):.3f}")

print(f"\n===   ===")
print(":")
print("    = 0,  = 1 ")
print("     0~1")
print("     ")
print("        ")

print("\n:")
print("    = 0,  = 1 ")
print("     ")
print("      ")
print("     (, 1   )")
````
